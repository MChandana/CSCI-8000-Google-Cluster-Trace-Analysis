```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import glob as glob
```

```python
task_event_df.columns=[ 'timestamp', 'task_id','cpu','mem']
task_event_df.shape
```

```python
task_usage_df=pd.concat( [pd.read_csv(f, compression="gzip", header=None, usecols=[0,1,3,5,6,10,13]) for f in glob.glob("C:\\Users\\so51713\\Documents\\GoogleCloudDS\\task_usage\\part-0000*.csv.gz")])
task_usage_df.columns=['start_time','end_time', 'task_id','mean_cpu','canonical_mem','cpu','mem']
task_usage_df.shape
```

```python
task_event_data=task_event_df
task_event_data=task_event_data[(task_event_data.timestamp<47700000000)]
task_event_data.shape
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
ax.plot(task_event_data['timestamp'], task_event_data['cpu'], label='cpu requested')
ax.plot(task_usage_df['start_time'], task_usage_df['cpu'], label='cpu usage')
plt.xticks(np.arange(0,1,step=0.01))

plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
ax.plot(task_event_data['timestamp'], task_event_data['mem'], label='memory requested')
ax.plot(task_usage_df['start_time'], task_usage_df['mem'], label='memory usage')
#plt.xlim(min(samples_df['timestamp']), max(samples_df['timestamp']))
plt.ylim(np.arange(0,100,step=10))
plt.legend()
plt.show()
```
